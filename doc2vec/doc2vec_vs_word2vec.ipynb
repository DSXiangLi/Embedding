```python
import os 
from config.doc2vec_config import CHECKPOINT_DIR, DV_PARAMS, WV_PARAMS
from doc2vec.model import Doc2VecModel, Word2VecModel, BaseModel
Root_dir = '/Users/xiangli/Desktop/Embedding/'
data = 'sogou_news_big'
data_path = os.path.join(Root_dir, 'data','sogou_news_big','corpus_new.txt')
model_dir = os.path.join(Root_dir, CHECKPOINT_DIR) 
```

## Sogou News: 对比word2vec,doc2vec在变长文本向量上的表达差异

- 用相同参数，在Sogou News数据集上分别训练doc2vec, word2vec模型, 这里doc2vec用PV-DM因为要得到词向量
- PV-DM得到的文本向量，在训练集和测试集(infer)上的差异
- PV-DM得到的词向量，和word2vec得到的词向量差异
- PV-DM得到的文本向量，和word2vec做central embedding得到的文本向量的差异

## Doc2vec & Word2vec 模型加载

```python
mm = Doc2VecModel(DV_PARAMS, model_dir = model_dir.format('doc2vec', data ),  
             data_path = data_path)
mm.model_load()
```

```python
nn = Word2VecModel(WV_PARAMS, model_dir = model_dir.format('word2vec', data),
                   data_path = data_path)
nn.model_load() 
```

### 对比 Doc2vec & word2vec 得到的词向量

```python
words = ['暴雨','大雨','雨','下雨','小雨','细雨']
for i in words:
    print('\nWord count for {} = {}'.format(i,nn.model.wv.vocab[i].count ))
    print('Doc2vec:', [ i for i,j in mm.model.wv.most_similar(positive=i,topn=10)]) 
    print('Word2vec:',[ i for i,j in nn.model.wv.most_similar(positive=i,topn=10)]) 
```
